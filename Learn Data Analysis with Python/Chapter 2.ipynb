##**Combining Data from Multiple Excel Files**

```python
##Listing 2-15
#the code format was changed from book due to attribute error
#the code by replacing the append method with pd.concat().
#Long Way
import pandas as pd
import numpy as np
all_data=pd.DataFrame()

df=pd.read_excel('data1.xlsx')
all_data=pd.concat([all_data, df], ignore_index=True)

df=pd.read_excel('data2.xlsx')
all_data=pd.concat([all_data, df], ignore_index=True)

df=pd.read_excel('data3.xlsx')
all_data=pd.concat([all_data, df], ignore_index=True)
display(all_data.describe())
```

```python
##Listing 2-16
#the code format was changed from book due to attribute error
#the code by replacing the append method with pd.concat().
#Short Way
import pandas as pd
import numpy as np
import glob

all_data=pd.DataFrame()
for f in glob.glob("data*.xlsx"):
    df=pd.read_excel(f)
    all_data=pd.concat([all_data, df], ignore_index=True)
all_data.describe()
```

## **Saving Data to SQL**

```python
##Listing 2-21
#Create Dataset to Save
import pandas as pd
names=['Bob','Jessica','Mary','John','Mel']
grades=[76,95,77,78,99]
GradeList=zip(names,grades)
df=pd.DataFrame(data=GradeList,columns=['Names','Grades'])
df
```

```python
##Listing 2-22
#Export Dataframe to sqlite
import os
import sqlite3 as sq3
db_filename=r'mydb.db'
con=sq3.connect(db_filename)
df.to_sql('mytable',
          con,
          #flavor='sqlite' ---> this command shows Typeerror
          schema=None,
          if_exists='replace',
index=True,
index_label=None,
chunksize=None,
dtype=None)
con.close()
```

##**Random Numbers and Creating Random Data**

```python
##Listing 2-24
#Getting Started
import pandas as pd
from numpy import random
from numpy.random import randint
names=['Bob','Jessica','Mary','John','Mel']
```

```python
##Listing 2-25
#Seeding Random Generator
random.seed(500)
```

```python
##Listing 2-26
#Selecting 1000 Random Names
randnames=[]
for x in range(1000):
    name=names[randint(low=0,high=len(names))]
    randnames.append(names)
```

```python
##Listing 2-27
#Selecting 1000 Random Numbers
births=[]
for x in range(1000):
    births.append(randint(low=0,high=1000))
```

```python
##Listing 2-28
#Creating Dataset from the List of Random Names and Numbers
BabyDataSet2=list(zip(randnames,births))
df=pd.DataFrame(data=BabyDataSet2,
                columns=['Names','Births'])
df
```
